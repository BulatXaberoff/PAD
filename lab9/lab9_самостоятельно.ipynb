```python
import random
import nltk
nltk.download('punkt') 
nltk.download('punkt_tab')
from nltk.util import bigrams
from nltk.tokenize import word_tokenize

# Пример текста для генерации (можно заменить на любой другой)
text = """
    В начале марта один из почтенных и уважаемых помещиков нашего уезда, Федор Павлович Карамазов, возвратился из Петербурга, где проживал долго, в свой уездный городок, где находилось и его поместье. Его поместье состояло из двух частей: одно большое, на берегу реки, с пахотными полями и лесом, другое меньшее, но очень красивое, в горах, с садом и виноградником. При возвращении его на родину никто не встречал, хотя и многие ждали; впрочем, кроме его собственных домашних. Его прислуга состояла из трех человек: повара и двух лакеев. Один из лакеев был, впрочем, уже стар и выслужил свой век; другой молодой, лет восемнадцати, но глуповатый, и, что хуже всего, пьяница. Главная же прислуга Федора Павловича была в то время его дворовый человек Смердяков. Но о Смердякове после. Федор Павлович Карамазов был помещиком и в то же время одним из самых больших пьяниц и скандалистов во всей нашей округе. И хотя он был не дурен собой, однако же нажил себе такую репутацию, что с ним никто не хотел водиться. Его все ненавидели и презирали, и не только помещики, но даже и крестьяне, которых он мучил.
    Он был очень богат, и его богатство было приобретено не честным путем. Он был хитрый и коварный человек, умевший пользоваться слабостями других. Он обманывал и грабил всех, кого только мог. Но, несмотря на все это, он был очень любим своими детьми, особенно двумя старшими: Дмитрием и Иваном. Младший же его сын, Алеша, был очень тих и кроток, и не любил отца. Федор Павлович Карамазов, как мы уже сказали, был очень богат, и этот факт был главной причиной его уважения в обществе.
"""

# Токенизация текста (разделение на слова)
words = word_tokenize(text.lower())  # Приводим все слова к нижнему регистру

# Создание биграмм (пары соседних слов)
bigrams_list = list(bigrams(words))

# Функция для генерации предложения на основе биграмм
def generate_sentence_from_bigram(bigrams_model, num_words=10):
    # Начинаем с случайной биграммы
    word_pair = random.choice(bigrams_model)
    sentence = list(word_pair)
    
    # Генерация последующих слов
    for _ in range(num_words - 2):
        # Ищем все биграммы, начинающиеся с последнего слова в предложении
        possible_words = [bigram[1] for bigram in bigrams_model if bigram[0] == sentence[-1]]
        
        # Выбираем случайное слово из возможных
        next_word = random.choice(possible_words)
        sentence.append(next_word)
    
    # Собираем предложение и возвращаем
    return ' '.join(sentence).capitalize() + '.'

# Генерация предложения
generated_sentence = generate_sentence_from_bigram(bigrams_list, num_words=10)
print("Сгенерированное предложение:", generated_sentence)
```

```python
import nltk
from nltk import ngrams
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import random
import string

def preprocess_text(text):
  text = text.lower()
  text = text.translate(str.maketrans('', '', string.punctuation))
  return text


def generate_sentence_from_bigrams(bigram_freq, start_token, max_length=100, min_sentence_length = 5):
    current_sentence = [start_token]

    for _ in range(max_length):
        last_word = current_sentence[-1] #последнее слово
        next_bigrams = [bi for bi in bigram_freq.keys() if bi[0] == last_word] #биграммы, начинающиеся с последнего слова
        
        if not next_bigrams:
            break # если биграмм не нашлось - заканчиваем генерацию
        
        next_bigram = random.choices(next_bigrams, weights = [bigram_freq[bi] for bi in next_bigrams])[0]
        current_sentence.append(next_bigram[1])  # выбираем следующее слово

        if current_sentence[-1] in ['.','?','!'] and len(current_sentence) >= min_sentence_length:
            break; # если встречен знак конца предложения - заканчиваем генерацию

    return " ".join(current_sentence)

def generate_sentences(filepath, num_sentences=3):
    try:
        with open(filepath, 'r', encoding='utf-8') as file:
            text = file.read()
    except FileNotFoundError:
        return "Файл не найден"

    text = preprocess_text(text)
    tokens = word_tokenize(text)

    if len(tokens) < 2:
        return "Недостаточно данных для генерации предложений"

    bigrams = list(ngrams(tokens, 2))
    bigram_freq = FreqDist(bigrams) #считаем частоты биграмм

    generated_sentences = []
    for _ in range(num_sentences):
        start_token = random.choice(tokens)  #выбираем случайное слово
        generated_sentences.append(generate_sentence_from_bigrams(bigram_freq, start_token))

    return generated_sentences

# Пример использования
nltk.download('punkt')  # Загрузка токенизатора
filepath = "brothers_karamazov.txt"  # Замените на путь к вашему файлу

# Пример создания файла brothers_karamazov.txt с текстом "Братьев Карамазовых"
with open(filepath, 'w', encoding='utf-8') as file:
  file.write("""
    В начале марта один из почтенных и уважаемых помещиков нашего уезда, Федор Павлович Карамазов, возвратился из Петербурга, где проживал долго, в свой уездный городок, где находилось и его поместье. Его поместье состояло из двух частей: одно большое, на берегу реки, с пахотными полями и лесом, другое меньшее, но очень красивое, в горах, с садом и виноградником. При возвращении его на родину никто не встречал, хотя и многие ждали; впрочем, кроме его собственных домашних. Его прислуга состояла из трех человек: повара и двух лакеев. Один из лакеев был, впрочем, уже стар и выслужил свой век; другой молодой, лет восемнадцати, но глуповатый, и, что хуже всего, пьяница. Главная же прислуга Федора Павловича была в то время его дворовый человек Смердяков. Но о Смердякове после. Федор Павлович Карамазов был помещиком и в то же время одним из самых больших пьяниц и скандалистов во всей нашей округе. И хотя он был не дурен собой, однако же нажил себе такую репутацию, что с ним никто не хотел водиться. Его все ненавидели и презирали, и не только помещики, но даже и крестьяне, которых он мучил.
    Он был очень богат, и его богатство было приобретено не честным путем. Он был хитрый и коварный человек, умевший пользоваться слабостями других. Он обманывал и грабил всех, кого только мог. Но, несмотря на все это, он был очень любим своими детьми, особенно двумя старшими: Дмитрием и Иваном. Младший же его сын, Алеша, был очень тих и кроток, и не любил отца. Федор Павлович Карамазов, как мы уже сказали, был очень богат, и этот факт был главной причиной его уважения в обществе.
  
  """)
# Конец создания файла
generated = generate_sentences(filepath)
if isinstance(generated, list):
  for i, sentence in enumerate(generated):
    print(f"Предложение {i+1}: {sentence}")
else:
  print(generated)
```
