```python
import matplotlib.pyplot as plt
import numpy as np

def draw_circle_with_sectors(sectors):
    # Проверка на корректность ввода
    if not (3 <= sectors <= 100):
        print("Количество секторов должно быть от 3 до 100.")
        return

    # Углы для секторов
    angles = np.linspace(0, 2 * np.pi, sectors + 1)
    
    # Палитра цветов rainbow
    colors = plt.cm.rainbow(np.linspace(0, 1, sectors))
    
    # Построение секторов
    fig, ax = plt.subplots()
    for i in range(sectors):
        x = [0, np.cos(angles[i]), np.cos(angles[i+1]), 0]
        y = [0, np.sin(angles[i]), np.sin(angles[i+1]), 0]
        ax.fill(x, y, color=colors[i], edgecolor='black')

    # Настройка отображения
    ax.set_aspect('equal')  # Равные оси для круга
    plt.title(f"Окружность с {sectors} секторами")
    plt.show()

# Ввод количества секторов
n = int(input("Введите количество секторов (от 3 до 100): "))
draw_circle_with_sectors(n)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_colored_circle(num_sectors):
    """
    Рисует окружность, разделенную на сектора с равномерным распределением цветов.
    :param num_sectors: количество секторов (от 3 до 100)
    """
    if not (3 <= num_sectors <= 100):
        print("Ошибка: количество секторов должно быть от 3 до 100.")
        return

    colors = plt.cm.rainbow(np.linspace(0, 1, num_sectors))

    fig, ax = plt.subplots()
    ax.axis('equal')  # Окружность должна быть пропорциональной

    wedges, _ = ax.pie(
        [1] * num_sectors,  # Равномерное распределение секторов
        colors=colors,      # Палитра цветов
        startangle=90       # Начальный угол (сверху)
    )

    plt.title(f"Окружность с {num_sectors} секторами", fontsize=12)

    plt.show()

if __name__ == "__main__":
    try:
        num_sectors = int(input("Введите количество секторов (от 3 до 100): "))
        draw_colored_circle(num_sectors)
    except ValueError:
        print("Ошибка: введите целое число от 3 до 100.")
```
