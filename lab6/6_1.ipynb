```python
import scipy.stats as stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Parameters
A = 1  # Mean (mathematical expectation)
sigma = 3  # Standard deviation

try:
		A = int(input("Введите математическое ожидание"))
		sigma = int(input("Введите среднее отклонение"))
		sigma1 = int(input("Введите 1 среднее отклонение"))
		sigma2 = int(input("Введите 2 среднее отклонение"))
except ValueError:
		print("Неверный ввод. Пожалуйста, введите целые числа.")

sigmas = [sigma,sigma1,sigma2]

for s in sigmas:
	# Generate x values for plotting
	x = np.linspace(A - 4 * s, A + 4 * s, 500)

	# Calculate probability density function (PDF) for normal distribution
	pdf = stats.norm.pdf(x, loc=A, scale=s)

	# Plot the PDF
	plt.figure(figsize=(10, 6))
	plt.plot(x, pdf, label=f'N({A}, {s}^2)', color='blue')
	plt.fill_between(x, pdf, alpha=0.2, color='blue')  # Highlighting PDF area

	# Step 2: Calculate probability for interval (-A/6, A)
	interval = (-A / 6, A)
	probability = stats.norm.cdf(interval[1], loc=A, scale=s) - stats.norm.cdf(interval[0], loc=A, scale=s)

	# Highlight the interval on the plot
	highlight_x = np.linspace(interval[0], interval[1], 100)
	highlight_pdf = stats.norm.pdf(highlight_x, loc=A, scale=s)
	plt.fill_between(highlight_x, highlight_pdf, color='orange', alpha=0.6, label=f'P(X in {interval})')

	# Add titles and labels
	plt.title(f"Нормальное распределение. Вероятность попадания в интервал: {round(probability,2)*100}%")
	plt.xlabel("X")
	plt.ylabel("Density")
	plt.legend()
	plt.grid()
	plt.show()





```

```python
import os
# Step 3: Generate (X, Y) values for most probable distribution
x_values = np.linspace(A - 3 * sigma, A + 3 * sigma, 20)  # 20 evenly spaced points
y_values = stats.norm.pdf(x_values, loc=A, scale=sigma)

# Save (X, Y) data to file
current_directory = os.getcwd()

# Путь для сохранения файла	
file_path = os.path.join(current_directory, "normal_distribution_data.xlsx")
pd.DataFrame({"X": x_values, "Y": y_values}).to_excel(file_path, index=False)
```
