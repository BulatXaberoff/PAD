```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function():
    # X in the range from 1 to 4
    X = np.linspace(-1, 4, 100)
    Y = 3 * X - X**2 + 1

    # Plot Y(X) = 3X - X^2 + 1
    plt.plot(X, Y, label="Y(X) = 3X - X^2 + 1")

    # Find and display Y(15)
    x_val = 15
    y_val = 3 * x_val - x_val**2 + 1
    print(f"Y(15) = {y_val}")

    # Calculate and plot tangent line at X = 15
    slope = 3 - 2 * x_val
    tangent_line = slope * (X - x_val) + y_val
    plt.plot(X, tangent_line, '--', label=f"Tangent at X=15")

    # Show plot
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_function()
```
